### concat

```python
d={"city":['Guntur','Vijayawada','Vizag'],"count":[120,150,130],"area":[300.23,500.43,450.67]}
df1=pd.DataFrame(d)
df1
```

```python
d2={"city":['Tirupati','Srikakulam','anathapur'],"count":[150,170,120],"area":[200.23,400.43,350.67]}
df2=pd.DataFrame(d2)
df2
```

```python
total_df=pd.concat([df1,df2])
total_df
```

```python
total_df=pd.concat([df1,df2],ignore_index=True)
total_df
```

```python
total_df=pd.concat([df1,df2],keys=['a','b'])
total_df
```

```python
total_df.loc['a']
```

```python
total_df=pd.concat([df1,df2],axis=1)
total_df
```

### merge/join

```python
d={"sno":[101,102,103,104],"sname":["a","b","c","d"]}
df1=pd.DataFrame(d)
df1
```

```python
f={"sno":[101,102,103,104,105,110],"address":["ab","bc","cd","dd",'ee','ff']}
df2=pd.DataFrame(f)
df2
```

```python
s=pd.merge(df1,df2,on='sno',how='right')
s
```

```python
s=pd.DataFrame({"sno":[101,102,103,104],"sname":["a","b","c","d"]})
s
```

```python
s2=pd.DataFrame({'sno':[101,102,103,110],"saddress":["srikakulam","kadapa","Tuni",'hyd']})
s2
```

```python
merge_df=pd.merge(s,s2,on='sno')
merge_df
```

```python
merge_df=pd.merge(s,s2,on='sno',how='outer')
merge_df
```

```python
merge_df=pd.merge(s,s2,on='sno',how='inner')
merge_df
```

```python
merge_df=pd.merge(s,s2,on='sno',how='right')
merge_df
```

```python
merge_df=pd.merge(s,s2,on='sno',how='left')
merge_df
```

```python
merge_df.sort_values('sno')
```

### Visulize the data using KDE

Keranal Dencity Estimate plot is used for display the multiple samples data into one graph.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.title("Before scaling data")
sns.kdeplot(df['TV'])
sns.kdeplot(df['radio'])
sns.kdeplot(df['newspaper'])
sns.kdeplot(df['sales'])
```

```python
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
std_data=std.fit_transform(df)
std_data
```

```python
std_data_df=pd.DataFrame(std_data,columns=df.columns)
std_data_df.head()
```

```python
plt.title("After scaling data")
sns.kdeplot(std_data_df['TV'])
sns.kdeplot(std_data_df['radio'])
sns.kdeplot(std_data_df['newspaper'])
sns.kdeplot(std_data_df['sales'])
```

### Robost Scalling

Robost scalling also used for scale the outliers, it scale using median and Inter Quartail Range(IQR)

```python
from sklearn.preprocessing import RobustScaler
rb=RobustScaler()
rb_data=rb.fit_transform(df)
rb_data
```

```python
rb_data_df=pd.DataFrame(rb_data,columns=df.columns)
rb_data_df
```

```python
plt.title("After scaling data in robost scaler")
sns.kdeplot(rb_data_df['TV'])
sns.kdeplot(rb_data_df['radio'])
sns.kdeplot(rb_data_df['newspaper'])
sns.kdeplot(rb_data_df['sales'])
```

### Range scaling(Min Max Scaller)

By using MinMaxScaller we can provide custom range for scaling the data

```python
from sklearn.preprocessing import MinMaxScaler
m=MinMaxScaler(feature_range=(-1,1))
m_data=m.fit_transform(df)
m_data
```

```python
m_data_df=pd.DataFrame(m_data,columns=df.columns)
m_data_df
```

```python
plt.title("After scaling data using range scaler")
sns.kdeplot(m_data_df['TV'])
sns.kdeplot(m_data_df['radio'])
sns.kdeplot(m_data_df['newspaper'])
sns.kdeplot(m_data_df['sales'])
```

### Normalization:

* upto now we are scale by using features(columns)
* In certain cases we want to scale the individual data observation(rows)
*  when clustering data we need to apply normalization

```python
from sklearn.preprocessing import Normalizer
n=Normalizer()
n_data=n.fit_transform(df)
n_data
```

```python
n_data_df=pd.DataFrame(n_data,columns=df.columns)
n_data_df.head()
```

```python
plt.title("After scaling data using normalize scaler")
sns.kdeplot(n_data_df['TV'])
sns.kdeplot(n_data_df['radio'])
sns.kdeplot(n_data_df['newspaper'])
sns.kdeplot(n_data_df['sales'])
```
