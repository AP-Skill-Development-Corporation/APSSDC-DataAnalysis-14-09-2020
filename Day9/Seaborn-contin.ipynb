```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df.head()
```

#### Regression Plot
- It creates a regression line between two parameters and helps to visulize their linear realtionship
- lmplot

```python
print(dir(sns))
```

```python
df.columns
```

```python
sns.lmplot(x="total_bill",y="tip",data=df)
```

```python
sns.lmplot(x="total_bill",y="tip",data=df,hue="sex",markers=["d","s"])
```

```python
sns.lmplot(x="total_bill",y="tip",data=df,hue="sex",markers=["d","s"],palette="BuPu_r",scatter_kws={"s":30})
```

```python
help(sns.lmplot)
```

```python
# Displaying Multiple plots
sns.lmplot(x="total_bill",y="tip",data=df,col="sex",row="time",hue="smoker")
```

```python
# Displaying Multiple plots
sns.lmplot(x="total_bill",y="tip",data=df,col="sex",row="time",hue="smoker",height=6,palette="coolwarm")
```

```python
# bin represent how many falls in data
sns.distplot(df["total_bill"],bins=20)
```
