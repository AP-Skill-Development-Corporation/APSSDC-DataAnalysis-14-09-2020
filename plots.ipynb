```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df=pd.read_csv('iris.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
# find the how many datapoints in easch class
df['Series_names'].value_counts()

#balanced data set : if features have equal datapoints then it is known as balanced dataset
#imbalanced dataset : if features have non equal datapoints then it is knows as imbalanced dataset
```

### 2-D scatter plot:

```python
df.plot(x='sl',y='sw',kind='scatter')
#df.plot('sl','sw',kind='scatter')

plt.show()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(df,hue='Series_names',size=5).map(plt.scatter,'sl','sw').add_legend()
plt.show()
```

### Pair-Plots

```python
sns.set_style("whitegrid")
sns.pairplot(df,hue='Series_names',size=4)
plt.show()
```

### 1d scatter plots: Histogram plot

```python
sns.FacetGrid(df,hue='Series_names',size=4) \
.map(sns.distplot,'pl') \
.add_legend();
plt.show()

# dark blue line is called probability dencity function(pdf) is known as smooth function
```

```python
sns.FacetGrid(df,hue='Series_names',size=5).map(sns.distplot,'pw').add_legend()
plt.show()
```

```python
sns.FacetGrid(df,hue='Series_names',size=5).map(sns.distplot,'sl').add_legend()
plt.show()
```

```python
sns.FacetGrid(df,hue='Series_names',size=5).map(sns.distplot,'sw').add_legend()
plt.show()
```

```python
c,b=np.histogram(df['pl'],bins=10,density=True)
pdf=c/sum(c)
print(pdf)
print(b)

cdf=np.cumsum(pdf)
plt.plot(b[1:],pdf)
plt.plot(b[1:],cdf)
plt.show()
```

```python
setosa=df[df['Series_names']=='Iris-setosa']
versicolor=df[df['Series_names']=='Iris-versicolor']
verginica=df[df['Series_names']=='Iris-virginica']
verginica.head()
```

```python
c,b=np.histogram(setosa['pl'],bins=10,density=True)
pdf=c/sum(c)
print(pdf)
print(b)

cdf=np.cumsum(pdf)
plt.plot(b[1:],pdf)
plt.plot(b[1:],cdf)


#versicolor cdf and pdf
c,b=np.histogram(verginica['pl'],bins=10,density=True)
pdf=c/sum(c)
print(pdf)
print(b)

cdf=np.cumsum(pdf)
plt.plot(b[1:],pdf)
plt.plot(b[1:],cdf)

#verginica of cdf and pdf
c,b=np.histogram(versicolor['pl'],bins=10,density=True)
pdf=c/sum(c)
print(pdf)
print(b)

cdf=np.cumsum(pdf)
plt.plot(b[1:],pdf)
plt.plot(b[1:],cdf)


plt.show()
```

### box plot:

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('iris.csv')
df.head()
```

```python
sns.boxplot(x='Series_names',y='pl',data=df)
plt.show()
```

### Violin plots:

The combination of box plot and histogram&pdf plot is known as violin plot

```python
sns.violinplot(x='Series_names',y='pl',data=df,size=5)
plt.show()
               
```
