### subplots:

syntax: plt.subplots(row,column,pos)


```python
plt.subplots(2,2) # 2 rows and 2 columns
plt.show()
```

```python
x=np.array([10,20,30,40])
y=np.array([50,60,70,80])
plt.subplot(2,2,1)  # 2 rows,2 columns and 1 position
plt.plot(x,y,'r*')

plt.subplot(2,2,2)
plt.plot(x,y,'g--')

plt.subplot(2,2,3)
plt.plot(x,y,'b-')

plt.subplot(2,2,4)
plt.plot(x,x*y,'yo')
```

```python
x=np.array([10,20,30,40])
y=np.array([50,60,70,80])
plt.subplot(1,1,1)  # 2 rows,2 columns and 1 position
plt.plot(x,y,'r*')

# plt.subplot(1,2,2)  # 2 rows,2 columns and 1 position
# plt.plot(x,y,'r*')
```

```python
x=np.array([2,4,6,8,10,12,14,16])
plt.plot(x,x**2,'b--',label="dashed line")
plt.plot(x,x**2.2,'rs',label='squared line')
plt.plot(x,x**2.4,'g^',label='traiangle')
plt.legend()
plt.grid()
plt.show()
```

### Bar plot

```python
x=np.array([10,20,30,40])
y=np.array([50,60,70,80])
plt.bar(x,y,color='g')
plt.show()
```

```python
x=np.array([10,20,30,40])
y=np.array([50,60,70,80])
plt.bar(x,y,width=5)
plt.show()
```

```python
x=['naga lakshmi','sarayu','siva','venkata sai','hema']
y=[89,90,100,79,80]

plt.bar(x,y)
plt.show()
```

```python
x=['naga lakshmi','sarayu','siva','venkata sai','hema']
y=[89,90,100,79,80]

plt.bar(x,y)

for a,b in zip(x,y):
    #plt.text(a,b,str(b))
    #print(b)
    plt.text(a,b,str(b))

plt.show()
```

```python
# display the bar graph in horizontal
x=['naga lakshmi','sarayu','siva','venkata sai','hema']
y=[89,90,100,79,80]
plt.barh(x,y)
for a,b in zip(y,x):
    plt.text(a,b,str(a))

plt.show()
```

```python
b=[20,35,30,35,27,40]
g=[25,32,34,15,20,32]

i=np.arange(0,6)

p1=plt.bar(i,b,width=0.5)
p2=plt.bar(i,g,width=0.5,bottom=b)

plt.xlabel("class rooms")
plt.ylabel("strenght")
plt.title("mana college")
plt.xticks(i,('c1','c2','c3','c4','c5','c6'))
plt.yticks(np.arange(0,90,10))

plt.legend((p1[0],p2[0]),('boys','girls'))
plt.show()
```

### Box plot:

* this is also knows as whisker plot
* it contains min value,max value, first quartile, median, third quartaile

```python
data=np.random.normal(100,20,200)
plt.boxplot(data)
plt.show()
```

```python
a=np.random.randint(50,100,10)
b=np.random.randint(40,90,10)
c=np.random.randint(30,100,10)

plt.boxplot([a,b,c],labels=['b1','b2','b3'])
plt.show()
```

### Pie plot

```python
labels=['python','c','c++','java']
size=[100,200,300,450]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
plt.pie(size,labels=labels,colors=colors,autopct="%1.2f%%")
plt.axis('equal')
plt.show()
```

```python
labels=['python','c','c++','java']
size=[100,200,300,450]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(1,0,0,0)
plt.pie(size,labels=labels,explode=explode,colors=colors,autopct="%1.2f%%")
plt.axis('equal')
plt.show()
```

### save plot

```python
labels=['python','c','c++','java']
size=[100,200,300,450]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(1,0,0,0)
plt.pie(size,labels=labels,explode=explode,colors=colors,autopct="%1.2f%%")  #ex: 234.56
plt.axis('equal')
plt.savefig("pieplot2.jpg") #d://folder/filename.png
plt.show()
```
