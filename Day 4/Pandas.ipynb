### DataFrames

```python
# Create a DataFrame

import pandas as pd
import numpy as np

df=pd.DataFrame()
type(df)
```

```python
# pass the data into dataframe
l=[1,2,3,4,5,6,7,8,9]
df=pd.DataFrame(l)
df
```

```python
l=[[1,2],[3,4],[5,6]]
df=pd.DataFrame(l)
df
```

```python
l=[[1,2],[3,4],[5,6]]
df=pd.DataFrame(l,columns=["one","two"])
df
```

```python
# pass the data into dataframe using numpy
a=np.array(["Haritha","Vekata sai","pavan kumar","sravya"])
df=pd.DataFrame(a,columns=['Name'])
df
```

```python
# pass the data into dataframe using list of tuples
l=[("sivaram krishna",30),("one",12),("myname",45)]
df=pd.DataFrame(l)
df.columns=["name","marks"]
df
```

```python
# pass the data into dataframe using dictionary
d={"sname":["hello","good","siva","naga","manam"],"age":[90,100,45,32,6]}
df=pd.DataFrame(d)
df.columns=["your name","your age"]
df
```

```python
# add a new column
df['marks']=[12,1,3,14,25]
df
```

```python
d={"sno":[101,102,103,104,105],"sname":['a','b','c','d','e'],"sage":[10,12,14,11,17]}
df=pd.DataFrame(d)
df
```

```python
d={"sno":[101,102,103,104,105],"sname":['a','b','c','d','e'],"sage":[10,12,14,11,17]}
df=pd.DataFrame(d,index=[1,2,3,4,5])
df
```

```python
df.set_index('sno',inplace=True)
```

```python
df
```

```python
df.reset_index(inplace=True)
df
```

```python
k={'fname':['a','b','c','d','e'],"lname":['sai','ganesh','vijay','raju','ravi'],'marks':[31,85,23,20,55]}
df=pd.DataFrame(k)
df
```

```python
r=[]
for i in df['marks']:
    if i>0 and i<=25:
        r.append("fail")
    elif i>25 and i<60:
        r.append("second")
    else:
        r.append("first")
df['sresults']=r
df
```

```python
# using single condition

#df['sresults']=='fail'

df[df['sresults']=='fail']
```

```python
df[df['marks']>=50]
```

```python
df[(df['marks']<22) & (df['sresults']=='fail')]
```

### read the data into dataframe using files

* csv
    * pd.read_csv("filename.csv")
* excel
    * pd.read_excel("filename.xlsx)
* json
    * pd.read_json("filename.json")
* html
    * pd.read_html("fileurl")
